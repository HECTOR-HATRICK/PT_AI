```python
import pandas as pd
import numpy as np
import os
import shutil
```

```python
shutil.copytree("splitcoords", "splitcoords2") #don't want to mess up original data
```

```python
df=df.rolling(window=10, axis=1).mean()
```

```python
for filename in os.listdir("splitcoords2/train/data"):
    df=pd.read_csv(f"splitcoords2/train/data/{filename}", index_col=0)
    df=df.rolling(window=10, axis=1).mean()
    for i in range(9):
        df.drop(str(i), axis=1, inplace=True)
    df.to_csv(f"splitcoords2/train/data/{filename}")
```

```python
for filename in os.listdir("splitcoords2/val/data"):
    df=pd.read_csv(f"splitcoords2/val/data/{filename}", index_col=0)
    df=df.rolling(window=10, axis=1).mean()
    for i in range(9):
        df.drop(str(i), axis=1, inplace=True)
    df.to_csv(f"splitcoords2/val/data/{filename}")
```
